```python
import pandas as pd
import json
```

```python
df = pd.read_excel("masterData.xlsx")
```

```python
df.head()
```

```python
indexed_df = df.set_index(["stateCode","stateName","year"])
```

```python
indexed_df.head()
```

```python
output = []
for state, sub_df in indexed_df.groupby(level=["stateCode","stateName"]):
    data_df = sub_df.reset_index()
    year_df = data_df.pivot(index="stateCode",columns='year', values='year')
    population_df = data_df.pivot(index="stateCode",columns='year', values='population')
    category_df = data_df.pivot(index="stateCode",columns='year', values='category')
    populationRanking_df = data_df.pivot(index="stateCode",columns='year', values='populationRanking')
    unemploymentRate_df = data_df.pivot(index="stateCode",columns='year', values='unemploymentRate')
    violentCrimeRate_df = data_df.pivot(index="stateCode",columns='year', values='violentCrimeRate')
    data = {"stateCode":state[0],
        "stateName":state[1],
        "year":year_df.values.tolist()[0],
        "population":population_df.values.tolist()[0],
        "category":category_df.values.tolist()[0],
        "populationRanking":populationRanking_df.values.tolist()[0],
        "unemploymentRate":unemploymentRate_df.values.tolist()[0],
        "violentCrimeRate":violentCrimeRate_df.values.tolist()[0]
       }
    output.append(data)
```

```python
output
```

```python
with open('chartData.json', 'w') as json_output:
    json.dump({"chartData":output}, json_output)
```
